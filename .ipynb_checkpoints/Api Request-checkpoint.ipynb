```python
import requests
import pandas as pd
import os
import time
```

```python
url = "https://v3.football.api-sports.io/teams?league=262&season=2021"

payload={}
headers = {
  'x-rapidapi-key': 'eb56877777615b432a07c13c6e99f93b',
  'x-rapidapi-host': 'v3.football.api-sports.io'
}

teams = requests.request("GET", url, headers=headers, data=payload)

print(teams.text)
```

```python
team_json = teams.json()
```

```python
teams_json_clean = team_json['response']
```

```python
df_teams = pd.DataFrame(teams_json_clean) 
df_teams
```

```python
team_info = pd.json_normalize(df_teams['team'])
team_info.rename(columns = {'id': 'Team ID'}, inplace= True)
team_info.rename(columns = {'name': 'Team Name'}, inplace= True)
team_info.rename(columns = {'code': 'Team Code'}, inplace= True)
team_info.rename(columns = {'founded': 'Team Foundation Date'}, inplace= True)
team_info.rename(columns = {'national': 'National'}, inplace= True)
team_info.rename(columns = {'logo': 'Team Logo'}, inplace= True)
team_info.rename(columns = {'country': 'Team Country'}, inplace= True)
team_info
```

```python
team_venue = pd.json_normalize(df_teams['venue'])
team_venue.rename(columns = {'id': 'Stadium ID'}, inplace= True)
team_venue.rename(columns = {'name': 'Stadium Name'}, inplace= True)
team_venue.rename(columns = {'address': 'Stadium Address'}, inplace= True)
team_venue.rename(columns = {'city': 'City'}, inplace= True)
team_venue.rename(columns = {'capacity': 'Capacity'}, inplace= True)
team_venue.rename(columns = {'surface': 'Surface Type'}, inplace= True)
team_venue.rename(columns = {'image': 'Stadium Image'}, inplace= True)
type(team_venue)
```

```python
final_teams_info = pd.concat([team_info, team_venue], axis= 1, join = 'inner')
```

```python
final_teams_info
```

```python
final_teams_info.to_excel('LigaMx_teams.xlsx', sheet_name='LigaMx_Teams', index=False)
```
